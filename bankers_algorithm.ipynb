```python
import os
import random

def safety_algo():
    safe_list = []
    work = available.copy()
    updating = True
    while updating == True:
        prev_len = len(safe_list)
        for i in range(num_of_processes):
            if "P" + str(i) not in safe_list and False not in [x >= y for x, y in zip(work, need[i])]:
                safe_list.append("P" + str(i))
                work = [i + j for i, j in zip(work, allocation[i])]
        if len(safe_list) == prev_len:
            updating = False
    return safe_list

def need_output(m_dict, num_proc, num_work):
    size_of_work = num_work * 5
    print('\u250C{0:\u2500^6}{0:\u2500^{1}}\u2510'.format('', size_of_work + 1))
    print('|{0:^6}|'.format(''), end = '')
    print('{0:^{1}}|'.format('Need', size_of_work))
    print('\u251C{0:\u2500^6}{0:\u2500^{1}}\u2524'.format('', size_of_work + 1))
    print('|{0:^6}|'.format(''), end = '')
    for w in range(num_work):
        if w < num_work - 1:
            print('{0:^4}|'.format(chr(ord('A') + w)), end = '')
        else: 
            print('{0:^5}|'.format(chr(ord('A') + w)))
    print('\u251C{0:\u2500^6}{0:\u2500^{1}}\u2524'.format('', size_of_work + 1))
    for i in range(num_proc):
        print('|{0:^6}|'.format('P{}'.format(i)), end = '')
        for w in range(num_work):
            if w < num_work - 1:
                print('{0:^4}|'.format(m_dict[i][2][w]), end = '')
            else: 
                print('{0:^5}|'.format(m_dict[i][2][w]))
    print('\u2514{0:\u2500^6}{0:\u2500^{1}}\u2518'.format('', size_of_work + 1))

while True:
    
    print("Programmed by: Rey Victor Mendillo")
    print("MP05 - Banker's Algorithm\n")

    num_of_processes = int(input("Enter No. of Processes: "))
    num_of_resources = int(input("Enter No. of Resources: "))
    resource_instances = [0] * num_of_resources
    print("Enter No. of Instances per Resource: ", end='')
    for j in range(num_of_resources):
        if j > 0:
            print("                                     ", end='')
        resource_instances[j] = int(input(f"{chr(65 + j)} - "))

    allocation = [[0 for i in range(num_of_resources)] for i in range(num_of_processes)]
    max = [[0 for i in range(num_of_resources)] for i in range(num_of_processes)]
    need = [[0 for i in range(num_of_resources)] for i in range(num_of_processes)]

    has_negative = True
    while has_negative == True:
        for i in range(num_of_processes):
            for j in range(num_of_resources):
                max[i][j] = random.randint(0, resource_instances[j])
                allocation[i][j] = random.randint(0, max[i][j])
                need[i][j] = max[i][j] - allocation[i][j]
        total_allocation = [sum(i) for i in zip(*allocation)]
        available = [i - j for i, j in zip(resource_instances, total_allocation)]
        for j in range(num_of_resources):
            if available[j] < 0:
                has_negative = True
                break
            else:
                has_negative = False

    # To be Deleted            
    print("\nAllocation:", allocation)
    print("Max:", max)
    print("Need:", need)
    print("Total Allocation:", total_allocation)
    print("Available:", available)

    print("\nNeed Matrix")
    need_output({i: [allocation[i], max[i], need[i]] for i in range(num_of_processes)}, num_of_processes, num_of_resources)
    safe_list = safety_algo()
    safe = lambda safe_list: len(safe_list) == num_of_processes
    print("\nSafe?", safe(safe_list))

    if safe(safe_list):
        print("The safe sequence is:", safe_list)

        print("\nRequest")
        request_process = int(input("Enter Process No.: "))
        if request_process in range(num_of_processes):
            request_instances = [0] * num_of_resources
            print("Enter No. of Instances per Resource: ", end='')
            for j in range(num_of_resources):
                if j > 0:
                    print("                                     ", end='')
                request_instances[j] = int(input(f"{chr(65 + j)} - "))

            if available >= request_instances and False not in [x >= y for x, y in zip(need[request_process], request_instances)]:
                for i in range(num_of_processes):
                    if i == request_process:
                        allocation[i] = [x + y for x, y in zip(allocation[i], request_instances)]
                        need[i] = [x - y for x, y in zip(need[i], request_instances)]
                        available = [i - j for i, j in zip(available, request_instances)]
                        break

                # To be Deleted
                print("\nAllocation:", allocation)
                print("Max:", max)
                print("Need:", need)
                print("Total Allocation:", total_allocation)
                print("Available:", available)

                print("\nNeed Matrix")
                need_output({i: [allocation[i], max[i], need[i]] for i in range(num_of_processes)}, num_of_processes, num_of_resources)
                safe_list = safety_algo()
                safe = lambda safe_list: len(safe_list) == num_of_processes
                print("\nSafe?", safe(safe_list))

                if safe(safe_list):
                    print("The safe sequence is:", safe_list)
            else:
                print("\nThe request cannot be granted, since request is less than the available or the request is greater than the need.")
        else:
            print("\nProcess does not exist.")
    else:
        print("\nRequest is not possible, since the sequence is not safe.")
    print()
    choice = input("Do you want to run again [y/n]: ")
    while choice.lower() not in ('y','n'):
        choice = input("Do you want to run again [y/n]: ")
    if choice.lower() == 'n':
        break
    print()
os.system("pause")
```
